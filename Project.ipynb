–ü—Ä–æ–≤–µ–¥–µ–º –ø–µ—Ä–≤–∏—á–Ω–æ–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü. –î–ª—è –∫–∞–∂–¥–æ–π —Ç–∞–±–ª–∏—Ü—ã –≤—ã–≤–µ–¥–µ–º –∫–æ–ª–∏—á–Ω—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º —Å—Ç—Ä—É–∫—Ç—Ä—É.

```python
def collect_db_structure(engine, tables):
    """
    –°–æ–±–∏—Ä–∞–µ—Ç —Å—Ç—Ä—É–∫—Ç—É—Ä—É —Ç–∞–±–ª–∏—Ü –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö –≤ –æ–¥–∏–Ω DataFrame.
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        engine  ‚Äî SQLAlchemy engine
        tables  ‚Äî –∏—Ç–µ—Ä–∏—Ä—É–µ–º—ã–π —Å–ø–∏—Å–æ–∫ –Ω–∞–∑–≤–∞–Ω–∏–π —Ç–∞–±–ª–∏—Ü
    
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        DataFrame —Å–æ —Å—Ç–æ–ª–±—Ü–∞–º–∏:
            table_name, rows_count, column_name, data_type
    """
    result = []

    for table in tables['table_name']:
        # --- –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ ---
        cnt = pd.read_sql(
            f"SELECT COUNT(*) AS rows_count FROM {table};", 
            engine
        ).iloc[0]["rows_count"]

        # --- –°—Ç–æ–ª–±—Ü—ã –∏ —Ç–∏–ø—ã ---
        struct = pd.read_sql(f"""
            SELECT 
                column_name, 
                data_type
            FROM information_schema.columns
            WHERE table_name = '{table}';
        """, engine)

        # –¥–æ–±–∞–≤–ª—è–µ–º –≤—Å—ë –≤ –∏—Ç–æ–≥–æ–≤—ã–π —Å–ø–∏—Å–æ–∫
        for _, row in struct.iterrows():
            result.append({
                "table_name": table,
                "rows_count": cnt,
                "column_name": row["column_name"],
                "data_type": row["data_type"]
            })

    return pd.DataFrame(result)


df_structure = collect_db_structure(engine, tables)
df_structure
```

```python
def find_possible_keys(df):
    """
    –ù–∞—Ö–æ–¥–∏—Ç –≤–æ–∑–º–æ–∂–Ω—ã–µ –∫–ª—é—á–∏ –¥–ª—è —Å–≤—è–∑–∏ —Ç–∞–±–ª–∏—Ü:
    - –∏—â–µ—Ç –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫ –≤ —Ä–∞–∑–Ω—ã—Ö —Ç–∞–±–ª–∏—Ü–∞—Ö
    - –≤—ã–≤–æ–¥–∏—Ç —Ç–∞–±–ª–∏—Ü—É —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º
    """

    # –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ –∏–º–µ–Ω–∏ –∫–æ–ª–æ–Ω–∫–∏
    grouped = (
        df.groupby("column_name")["table_name"]
        .apply(list)
        .reset_index()
    )

    # –û—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ —Ç–µ –∫–æ–ª–æ–Ω–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –±–æ–ª–µ–µ —á–µ–º –≤ 1 —Ç–∞–±–ª–∏—Ü–µ
    repeated = grouped[grouped["table_name"].apply(len) > 1].copy()

    # –†–∞—Å–ø–∞–∫–æ–≤—ã–≤–∞–µ–º –≤ –¥–ª–∏–Ω–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç
    rows = []
    for _, row in repeated.iterrows():
        for table in row["table_name"]:
            rows.append({
                "column_name": row["column_name"],
                "table_name": table
            })

    return pd.DataFrame(rows)
dupes = find_possible_keys(df_structure)
dupes
```

### –ê–Ω–∞–ª–∏–∑ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏

–ü—Ä–æ–≤–µ–¥–µ–º –∞–Ω–∞–ª–∏–∑ –≤ –∫–∞–∫–∏–µ –º–µ—Å—è—Ü—ã –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –∞–≤–∞—Ä–∏–π. –î–ª—è –Ω–∞—á–∞–ª–∞ –ø–æ—Å–º–æ—Ç—Ä–∏–º —Å–∫–æ–ª—å–∫–æ –¥–∞–Ω–Ω—ã—Ö —Å–æ–¥–µ—Ä–∂–∏—Ç —Ç–∞–±–ª–∏—Ü–∞ `collisions` –ø–æ –≥–æ–¥–∞–º –∏ –º–µ—Å—è—Ü–∞–º.

```python
# --- SQL –∑–∞–ø—Ä–æ—Å ---
query = """
SELECT
    EXTRACT(YEAR FROM collision_date)  AS year,
    EXTRACT(MONTH FROM collision_date) AS month,
    COUNT(*) AS accidents_count
FROM collisions
GROUP BY year, month
ORDER BY year, month;
"""

# --- –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö ---
df_monthly = pd.read_sql(query, engine)

# --- –°–æ–∑–¥–∞–µ–º —É–¥–æ–±–Ω—É—é –∫–æ–ª–æ–Ω–∫—É –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ ---
df_monthly["period"] = df_monthly["year"].astype(int).astype(str) + "-" + df_monthly["month"].astype(int).astype(str).str.zfill(2)

# --- –ì—Ä–∞—Ñ–∏–∫ ---
plt.figure(figsize=(14, 6))
plt.bar(df_monthly["period"], df_monthly["accidents_count"])
plt.xticks(rotation=90)
plt.title("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –î–¢–ü –ø–æ –º–µ—Å—è—Ü–∞–º")
plt.xlabel("–ì–æ–¥-–º–µ—Å—è—Ü")
plt.ylabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–≤–∞—Ä–∏–π")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

–°—É–¥—è –ø–æ –≥—Ä–∞—Ñ–∏–∫—É, –±–∞–∑–∞ –ø–æ–∫—Ä—ã–≤–∞–µ—Ç –Ω–µ –≤–µ—Å—å –Ω—É–∂–Ω—ã–π –ø–µ—Ä–∏–æ–¥.
–î–æ –∏—é–Ω—è 2012 –≤—Å—ë –Ω–æ—Ä–º–∞–ª—å–Ω–æ ‚Äî –î–¢–ü –∑–∞–ø–∏—Å–∞–Ω—ã –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü.
–ü–æ—Å–ª–µ –∏—é–Ω—è 2012 –ø–æ—á—Ç–∏ –ø—É—Å—Ç–æ, –ø–æ—ç—Ç–æ–º—É –±–∞–∑–∞ –≤—ã–≥–ª—è–¥–∏—Ç –æ–±—Ä–µ–∑–∞–Ω–Ω–æ–π –∏–ª–∏ –Ω–µ–ø–æ–ª–Ω–æ–π.         
–î–∞–ª—å–Ω–µ–π—à–∏–µ –≤—ã–≤–æ–¥—ã –ø–æ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏ –∏ –¥–∏–Ω–∞–º–∏–∫–µ –Ω—è –±—É–¥—É –¥–µ–ª–∞—Ç—å –ø–æ –±–∞–∑–µ –¥–æ 2012 –≥–æ–¥–∞, —Ç–∞–∫ –∫–∞–∫ 2012 –≥–æ–¥ - –Ω–µ –ø–æ–ª–Ω—ã–π –∏ –¥–∞–Ω–Ω—ã–µ –ø–æ–ª—É—á–∞—Ç—Å—è –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–º–∏. –ë—Ä–∞—Ç—å –±—É–¥—É —Ç–æ–ª—å—É–æ –ø–æ–ª–Ω—ã–µ –≥–æ–¥–∞, —Ç–æ –µ—Å—Ç—å –∑–∞ 2009 - 2011 –≥.

```python
# SQL –∑–∞–ø—Ä–æ—Å
query = """
SELECT 
    DATE_TRUNC('month', collision_date)::date AS month,
    COUNT(*) AS accidents_count
FROM collisions
WHERE collision_date::date < '2012-01-01'
    AND collision_date::date >= '2009-01-01'
GROUP BY month
ORDER BY month;
"""

# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ –±–∞–∑—ã
df_month = pd.read_sql(query, engine)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∫–æ–ª–æ–Ω–∫—É –≤ –¥–∞—Ç—É
df_month['month'] = pd.to_datetime(df_month['month'], utc=True)
# –ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
min_val = df_month['accidents_count'].min()

# –ì—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(14,6))
plt.axhline(min_val, color='red', linestyle='--', linewidth=2, label=f"–ú–∏–Ω–∏–º—É–º: {min_val}")
plt.bar(df_month['month'].dt.strftime('%Y-%m'), df_month['accidents_count'])
plt.title("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –î–¢–ü –ø–æ –º–µ—Å—è—Ü–∞–º 2009-2011 –≥–≥")
plt.xlabel("–ú–µ—Å—è—Ü")
plt.ylabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–≤–∞—Ä–∏–π")
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

–í–∏–∑—É–∞–ª—å–Ω–æ –ª–∏–¥–∏—Ä—É–µ—Ç –æ–∫—Ç—è–±—Ä—å.
–°–¥–µ–ª–∞—é —Ç–∞–∫ –∂–µ –∞–Ω–∞–ª–∏–∑ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –º–µ—Å—è—Ü–∞–º, —á—Ç–æ–±—ã –≤—ã—è–≤–∏—Ç—å —Å–∞–º—ã–µ –∞–≤–∞—Ä–∏–π–Ω—ã–µ –∏ —Å–∞–º—ã–µ "–±–µ–∑–æ–ø–∞—Å–Ω—ã–µ" –º–µ—Å—è—Ü—ã.

```python
# SQL –∑–∞–ø—Ä–æ—Å ‚Äî —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–≤–∞—Ä–∏–π –ø–æ –∫–∞–∂–¥–æ–º—É –º–µ—Å—è—Ü—É
query = """
SELECT
    EXTRACT(MONTH FROM month_date) AS month,
    AVG(cnt) AS avg_accidents
FROM (
    SELECT
        DATE_TRUNC('month', collision_date)::date AS month_date,
        COUNT(*) AS cnt
    FROM collisions
    WHERE collision_date::date < '2012-01-01'
        AND collision_date::date >= '2009-01-01'
    GROUP BY month_date
) AS monthly
GROUP BY month
ORDER BY month;
"""

# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
df_avg = pd.read_sql(query, engine)

# –ì—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(12,6))
plt.bar(df_avg['month'], df_avg['avg_accidents'], color='skyblue')

plt.title("–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –î–¢–ü –ø–æ –º–µ—Å—è—Ü–∞–º (2009‚Äì2011)", fontsize=14)
plt.xlabel("–ú–µ—Å—è—Ü")
plt.ylabel("–°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –î–¢–ü")
plt.grid(axis='y', alpha=0.3)

plt.xticks(range(1,13))
plt.show()


# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã
plt.figure(figsize=(14, 2))
heat_data = df_avg[['avg_accidents']].T
heat_data.columns = df_avg['month']

sns.heatmap(
    heat_data,
    annot=True,
    fmt=".0f",
    cmap="YlOrRd",
    cbar=True
)

plt.title("–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –î–¢–ü –ø–æ –º–µ—Å—è—Ü–∞–º (2009‚Äì2011)")
plt.ylabel("")
plt.xlabel("–ú–µ—Å—è—Ü")
plt.show()
```

#### –ó–∞–¥–∞—á–∞ ‚Ññ1.

–ó–∞–¥–∞—á–∞ ‚Ññ1. –ö–∞–∫ –≤–ª–∏—è–µ—Ç —Å–æ—Å—Ç–æ—è–Ω–∏–µ –¥–æ—Ä–æ–≥–∏ –Ω–∞ –≤–∏–Ω–æ–≤–Ω–æ—Å—Ç—å –≤–æ–¥–∏—Ç–µ–ª—è?

–î–ª—è —Ä–µ—à–µ–Ω–∏—è —ç—Ç–æ–π –∑–∞–¥–∞—á–∏ –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã —Ç–∞–±–ª–∏—Ü—ã `collisions` —Å—Ç–æ–ª–±–µ—Ü `ROAD_SURFACE` –∏ `parties` —Å—Ç–æ–ª–±–µ—Ü `AT_FAULT`.                 
–í–∏–Ω–æ–≤–Ω–æ—Å—Ç—å –≤–æ–¥–∏—Ç–µ–ª—è - –±–∏–Ω–∞—Ä–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫:
- 0 - –Ω–µ –≤–∏–Ω–æ–≤–µ–Ω
- 1 - –≤–∏–Ω–æ–≤–µ–Ω

–°–æ—Å—Ç–æ—è–Ω–∏–µ –¥–æ—Ä–æ–≥–∏: 
- A ‚Äî Dry (–°—É—Ö–∞—è)
- B ‚Äî Wet (–ú–æ–∫—Ä–∞—è)
- C ‚Äî Snowy or Icy (–ó–∞—Å–Ω–µ–∂–µ–Ω–Ω–∞—è –∏–ª–∏ –æ–±–ª–µ–¥–µ–Ω–µ–ª–∞—è)
- D ‚Äî Slippery (Muddy, Oily, etc.) (–°–∫–æ–ª—å–∑–∫–∞—è, –≥—Ä—è–∑–Ω–∞—è, –º–∞—Å–ª—è–Ω–∏—Å—Ç–∞—è –∏ —Ç. –¥.)
- _ ‚Äî Not Stated (–ù–µ —É–∫–∞–∑–∞–Ω–æ)

```python
query = """
SELECT
    c.road_surface,
    AVG(p.at_fault) AS fault_rate,
    COUNT(*) AS total_cases
FROM collisions c
JOIN parties p
    ON c.case_id = p.case_id
  AND c.road_surface <> '-'
GROUP BY c.road_surface
ORDER BY fault_rate DESC;
"""

# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
df_1 = pd.read_sql(query, engine)

# –ì—Ä–∞—Ñ–∏–∫

plt.figure(figsize=(10,5))
plt.bar(df_1['road_surface'], df_1['fault_rate'], color='tomato')
plt.title("–î–æ–ª—è –≤–∏–Ω–æ–≤–Ω–æ—Å—Ç–∏ –≤–æ–¥–∏—Ç–µ–ª–µ–π –ø–æ —Å–æ—Å—Ç–æ—è–Ω–∏—é –¥–æ—Ä–æ–≥–∏")
plt.ylabel("–°—Ä–µ–¥–Ω—è—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤–∏–Ω–æ–≤–Ω–æ—Å—Ç–∏")
plt.xlabel("–°–æ—Å—Ç–æ—è–Ω–∏–µ –¥–æ—Ä–æ–≥–∏")
plt.grid(axis='y')
plt.show()
```

```python
query = """
SELECT 
    c.road_surface,
    AVG(CASE WHEN p.at_fault = 1 THEN 1 ELSE 0 END) AS fault_rate,
    AVG(CASE WHEN p.at_fault = 0 THEN 1 ELSE 0 END) AS not_fault_rate
FROM collisions c
JOIN parties p
    ON c.case_id = p.case_id
WHERE c.road_surface IS NOT NULL
GROUP BY c.road_surface
ORDER BY c.road_surface;
"""

df_fault = pd.read_sql(query, engine)

# –ì—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(12,6))
plt.plot(df_fault['road_surface'], df_fault['fault_rate'], marker='o', label='–î–æ–ª—è –≤–∏–Ω–æ–≤–Ω—ã—Ö')
plt.plot(df_fault['road_surface'], df_fault['not_fault_rate'], marker='o', label='–î–æ–ª—è –Ω–µ–≤–∏–Ω–æ–≤–Ω—ã—Ö')

plt.title("–î–æ–ª—è –≤–∏–Ω–æ–≤–Ω—ã—Ö –∏ –Ω–µ–≤–∏–Ω–æ–≤–Ω—ã—Ö —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤ –ø–æ —Å–æ—Å—Ç–æ—è–Ω–∏—é –¥–æ—Ä–æ–≥–∏")
plt.xlabel("–°–æ—Å—Ç–æ—è–Ω–∏–µ –¥–æ—Ä–æ–≥–∏ (ROAD_SURFACE)")
plt.ylabel("–î–æ–ª—è")
plt.grid(True)
plt.legend()
plt.show()
```

#### –ó–∞–¥–∞—á–∞ ‚Ññ2.

–ó–∞–¥–∞—á–∞ ‚Ññ2. –ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Å–µ—Ä—å—ë–∑–Ω–æ—Å—Ç–∏ –î–¢–ü –æ—Ç –≤–æ–∑—Ä–∞—Å—Ç–∞ –∞–≤—Ç–æ–º–æ–±–∏–ª—è

–¢–∞–±–ª–∏—Ü—ã: `collisions` –∏ `vehicles`
–ö–ª—é—á–∏: `case_id` + `party_number`

–î–ª—è —Ä–µ—à–µ–Ω–∏—è —ç—Ç–æ–π –∑–∞–¥–∞—á–∏ –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã —Ç–∞–±–ª–∏—Ü—ã `collisions` —Å—Ç–æ–ª–±–µ—Ü `collision_damage` –∏ `vehicles` —Å—Ç–æ–ª–±–µ—Ü `vehicle_age`.                 
–í–æ–∑—Ä–∞—Å—Ç –∞–≤—Ç–æ–º–æ–±–∏–ª—è:
- —á–∏—Å–ª–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –≥–æ–¥–∞—Ö

–°–µ—Ä—å—ë–∑–Ω–æ—Å—Ç—å –ø—Ä–æ–∏—Å—à–µ—Å—Ç–≤–∏—è–∏: 
- FATAL –¢–° (–ù–µ –ø–æ–¥–ª–µ–∂–∏—Ç –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏—é) 
- SEVERE DAMAGE (–°–µ—Ä—å—ë–∑–Ω—ã–π —Ä–µ–º–æ–Ω—Ç, –±–æ–ª—å—à–∞—è —á–∞—Å—Ç—å –ø–æ–¥ –∑–∞–º–µ–Ω—É/–°–µ—Ä—å—ë–∑–Ω–æ–µ –ø–æ–≤—Ä–µ–∂–¥–µ–Ω–∏–µ –∫–∞–ø–∏—Ç–∞–ª—å–Ω–æ–≥–æ —Å—Ç—Ä–æ–µ–Ω–∏—è) 
- MIDDLE DAMAGE (–°—Ä–µ–¥–Ω–∏–π —Ä–µ–º–æ–Ω—Ç, –º–∞—à–∏–Ω–∞ –≤ —Ü–µ–ª–æ–º –Ω–∞ —Ö–æ–¥—É/–°—Ç—Ä–æ–µ–Ω–∏–µ –≤ —Ü–µ–ª–æ–º —É—Å—Ç–æ—è–ª–æ) 
- SMALL DAMAGE (–û—Ç–¥–µ–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –∫—É–∑–æ–≤–∞ –ø–æ–¥ –∑–∞–º–µ–Ω—É/–ø–æ–∫—Ä–∞—Å–∫—É) 
- SCRATCH (–¶–∞—Ä–∞–ø–∏–Ω–∞)

–¢–∞–∫ –∫–∞–∫ —Å–µ—Ä—å—ë–∑–Ω–æ—Å—Ç—å –∞–≤–∞—Ä–∏–∏ —Ö—Ä–∞–Ω–∏—Ç—Å—è –≤ –≤–∏–¥–µ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π, —è –ø–µ—Ä–µ–≤–µ–¥—É –∫–∞–∂–¥—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é –≤ —á–∏—Å–ª–æ–≤–æ–π –±–∞–ª–ª. –¥–∞–ª–µ–µ –≤–æ–∑—å–º—É —Å—Ä–µ–¥–Ω–∏–π –±–∞–ª –ø–æ –≤–æ–∑—Ä–∞—Å—Ç—É –º–∞—à–∏–Ω—ã, –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏. 

–ß–µ–º —Å–µ—Ä—å–µ–∑–Ω–µ–µ –ø–æ–≤—Ä–µ–∂–¥–µ–Ω–∏–µ - —Ç–µ–º –≤—ã—à–µ –±–∞–ª–ª:
| –ö–∞—Ç–µ–≥–æ—Ä–∏—è     | –ë–∞–ª–ª |
| ------------- | ---- |
| SCRATCH       | 1    |
| SMALL DAMAGE  | 2    |
| MIDDLE DAMAGE | 3    |
| SEVERE DAMAGE | 4    |
| FATAL –¢–°      | 5    |


```python
query = """
SELECT 
    v.vehicle_age,
    AVG(
        CASE 
            WHEN c.collision_damage = 'fatal tc' THEN 5
            WHEN c.collision_damage = 'severe damage' THEN 4
            WHEN c.collision_damage = 'middle damage' THEN 3
            WHEN c.collision_damage = 'small damage' THEN 2
            WHEN c.collision_damage = 'scratch' THEN 1
            ELSE NULL
        END
    ) AS avg_damage_score,
    COUNT(*) AS n
FROM vehicles v
JOIN collisions c
    ON v.case_id = c.case_id
WHERE v.vehicle_age IS NOT NULL
  AND c.collision_damage IS NOT NULL
GROUP BY v.vehicle_age
HAVING COUNT(*) > 20
ORDER BY v.vehicle_age;
"""

df_2 = pd.read_sql(query, engine)

plt.figure(figsize=(12,5))
plt.plot(df_2['vehicle_age'], df_2['avg_damage_score'], marker='o')
plt.title("–°—Ä–µ–¥–Ω—è—è —Å–µ—Ä—å—ë–∑–Ω–æ—Å—Ç—å –î–¢–ü –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≤–æ–∑—Ä–∞—Å—Ç–∞ –∞–≤—Ç–æ–º–æ–±–∏–ª—è")
plt.xlabel("–í–æ–∑—Ä–∞—Å—Ç –∞–≤—Ç–æ–º–æ–±–∏–ª—è (–ª–µ—Ç)")
plt.ylabel("–°—Ä–µ–¥–Ω—è—è —Å–µ—Ä—å—ë–∑–Ω–æ—Å—Ç—å –ø–æ–≤—Ä–µ–∂–¥–µ–Ω–∏–π")
plt.grid(True)
plt.show()
```

```python
def show_df_info(df, name='DataFrame'):
    """
    –û—Ç–æ–±—Ä–∞–∂–∞–µ—Ç –æ—Å–Ω–æ–≤–Ω—ã–µ —Å–≤–µ–¥–µ–Ω–∏—è –æ –¥–∞—Ç–∞—Å–µ—Ç–µ:
    - –ø–µ—Ä–≤—ã–µ —Å—Ç—Ä–æ–∫–∏
    - –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ç–∏–ø–∞—Ö –¥–∞–Ω–Ω—ã—Ö –∏ –ø—Ä–æ–ø—É—Å–∫–∞—Ö
    - –æ–ø–∏—Å–∞—Ç–µ–ª—å–Ω—É—é —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É
    """
    buffer = StringIO()
    
    buffer.write(f'\n=== {name} ===\n')

    buffer.write('\n–ü–µ—Ä–≤—ã–µ —Å—Ç—Ä–æ–∫–∏:\n')
    buffer.write(df.head().to_string())
    buffer.write('\n')

    buffer.write('\n–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö:\n')
    df.info(buf=buffer)        
    buffer.write('\n')

    buffer.write('\n–û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:\n')
    buffer.write(df.describe(include="all").to_string())
    buffer.write('\n')

    print(buffer.getvalue())

show_df_info(crash, 'crash')
```

#### –ö–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏

```python
# –§—É–Ω–∫—Ü–∏—è: —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –ø—Ä–∏–∑–Ω–∞–∫ –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–º (–≤ —Ç.—á. float)
def is_discrete_feature(series, threshold=15):
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç –ø—Ä–∏–∑–Ω–∞–∫ –Ω–∞ –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ—Å—Ç—å
    –í —Ç–æ–º —á–∏—Å–ª–µ –µ—Å–ª–∏ —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö float.
    –ï—Å–ª–∏ –¥–∞–Ω–Ω—ã–µ –∏–º–µ—é—Ç –º–µ–Ω—å—à–µ 15 –∑–Ω–∞—á–µ–Ω–∏–π, —Ç–æ –ø—Ä–∏–∑–Ω–∞–∫ –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–π.
    """
    return (
        series.nunique() < threshold and
        (is_integer_dtype(series) or series.dropna().apply(float.is_integer).all())
    )

# –°–ª–æ–≤–∞—Ä—å —á–∏—Ç–∞–µ–º—ã—Ö –Ω–∞–∑–≤–∞–Ω–∏–π –æ—Å–µ–π 
axis_labels = {
    'insurance_premium': '–°—Ç—Ä–∞—Ö–æ–≤–∫–∞, —Ç—ã—Å.$',
    'week_day': '–î–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –∞–≤–∞—Ä–∏–∏',
    'month': '–ú–µ—Å—è—Ü –∞–≤–∞—Ä–∏–∏',
    'day': '–î–µ–Ω—å –º–µ—Å—è—Ü–∞ –∞–≤–∞—Ä–∏–∏',
    'hour': '–ß–∞—Å –∞–≤–∞—Ä–∏–∏',
    'vehicle_age': '–í–æ–∑–æ—Ä–∞—Å—Ç –∞–≤—Ç–æ–º–æ–±–∏–ª—è, –ª–µ—Ç'
}

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π ---
def plot_feature_distributions(df):
    """
    –°—Ç—Ä–æ–∏—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –±–æ–∫—Å–ø–ª–æ—Ç—ã –¥–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö.
    –î–ª—è –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–æ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ —Å–æ–∑–¥–∞—ë—Ç –¥–≤–∞ –≥—Ä–∞—Ñ–∏–∫–∞:
    1. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (–≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –∏–ª–∏ countplot –¥–ª—è –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤)
    2. –ë–æ–∫—Å–ø–ª–æ—Ç –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –≤—ã–±—Ä–æ—Å–æ–≤ –∏ —Ä–∞–∑–±—Ä–æ—Å–∞    
    
    """
    num_cols = df.select_dtypes(include='number').columns
    num_cols = [col for col in num_cols if col in axis_labels.keys()]

    for col in num_cols:
        label = axis_labels.get(col, col)
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))

        if is_discrete_feature(df[col]):
            sns.countplot(x=df[col], ax=axes[0], palette='viridis')
            axes[0].set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (–¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–µ):\n{label}')
            axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')
            axes[0].set_xlabel(label)
            axes[0].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–≤–∞—Ä–∏–π')
        else:
            sns.histplot(df[col], bins=20, kde=True, ax=axes[0], color='steelblue')
            axes[0].set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:\n{label}')
            axes[0].set_xlabel(label)
            axes[0].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–≤–∞—Ä–∏–π')

        sns.boxplot(x=df[col], ax=axes[1], color='orange')
        axes[1].set_title(f'–ë–æ–∫—Å–ø–ª–æ—Ç:\n{label}')
        axes[1].set_xlabel(label)

        plt.tight_layout()
        plt.show()


plot_feature_distributions(crash)
```

–í–æ–∑—Ä–∞—Å—Ç –∞–≤—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª—ë–Ω –Ω–µ—Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ: –≤ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–µ –î–¢–ü —É—á–∞—Å—Ç–≤—É—é—Ç –º–∞—à–∏–Ω—ã 3‚Äì7 –ª–µ—Ç.

#### –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏

```python
def plot_categorical_features(df):
    """
    –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.
    –ï—Å–ª–∏ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π > 10 ‚Äî –≤—ã–≤–æ–¥–∏—Ç—Å—è —Ç–æ–ª—å–∫–æ TOP-10.
    """

    axis_labels = {
        'intersection': '–Ø–≤–ª—è–µ—Ç—Å—è –ª–∏ –º–µ—Å—Ç–æ –ø—Ä–æ–∏—Å—à–µ—Å—Ç–≤–∏—è –ø–µ—Ä–µ–∫—Ä—ë—Å—Ç–∫–æ–º',
        'weather': '–ü–æ–≥–æ–¥–Ω—ã–µ —É—Å–ª–æ–≤–∏—è',
        'road_surface': '–°–æ—Å—Ç–æ—è–Ω–∏–µ –¥–æ—Ä–æ–≥–∏',
        'lighting': '–û—Å–≤–µ—â–µ–Ω–∏–µ',
        'control_device': '–ù–∞–ª–∏—á–∏–µ —Å–≤–µ—Ç–æ—Ñ–æ—Ä–∞',
        'direction': '–ù–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –¥–≤–∏–∂–µ–Ω–∏—è',
        'location_type': '–¢–∏–ø –º–µ—Å—Ç–æ–ø–æ–ª–æ–∂–µ–Ω–∏—è',
        'vehicle_type': '–¢–∏–ø –∞–≤—Ç–æ–º–æ–±–∏–ª—è',
        'vehicle_transmission': '–¢–∏–ø —Ç—Ä–∞–Ω—Å–º–∏—Å—Å–∏–∏ –∞–≤—Ç–æ–º–æ–±–∏–ª—è',
        'at_fault': '–í–∏–Ω–æ–≤–µ–Ω –ª–∏ –≤–æ–¥–∏—Ç–µ–ª—å'
    }

    # --- –í—ã–±–∏—Ä–∞–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ ---
    cat_cols = []

    for col in df.columns:
        # 1. –¢–∏–ø object ‚Üí –∫–∞—Ç–µ–≥–æ—Ä–∏—è
        if df[col].dtype == 'object':
            cat_cols.append(col)
            continue

        # 2. –ß–∏—Å–ª–æ–≤—ã–µ —Å –º–∞–ª—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π ‚Üí –∫–∞—Ç–µ–≥–æ—Ä–∏—è
        if df[col].nunique() <= 4:
            cat_cols.append(col)

    # --- –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ ---
    for i in range(0, len(cat_cols), 2):
        n_subplots = min(2, len(cat_cols) - i)
        fig, axes = plt.subplots(1, n_subplots, figsize=(16, 5))

        if n_subplots == 1:
            axes = [axes]

        for j in range(n_subplots):
            col = cat_cols[i + j]
            label = axis_labels.get(col, col)

            series = df[col]

            # === –û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ TOP-10 ===
            if series.nunique() > 10:
                top10 = series.value_counts().nlargest(10)
                # —Å–æ–∑–¥–∞—ë–º –Ω–æ–≤—É—é —Å–µ—Ä–∏—é: –≤—Å—ë –≤–Ω–µ —Ç–æ–ø–∞ ‚Üí "Other"
                series = series.apply(lambda x: x if x in top10.index else "Other")

            ax = sns.countplot(x=series, ax=axes[j], palette='crest')

            axes[j].set_xticklabels(axes[j].get_xticklabels(), rotation=45)
            axes[j].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
            axes[j].set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:\n{label}')

            # –ø–æ–¥–ø–∏—Å–∏ –Ω–∞–¥ —Å—Ç–æ–ª–±—Ü–∞–º–∏
            for p in ax.patches:
                height = p.get_height()
                ax.annotate(f'{height}',
                            (p.get_x() + p.get_width() / 2., height),
                            ha='center', va='bottom', fontsize=9)

        plt.tight_layout()
        plt.show()
plot_categorical_features(crash)
```

- –î–æ–∂–¥—å, –æ–±–ª–∞—á–Ω–æ—Å—Ç—å –∏ —Å–Ω–µ–≥ –¥–∞—é—Ç –Ω–∞–º–Ω–æ–≥–æ –º–µ–Ω—å—à–µ —Å–ª—É—á–∞–µ–≤.
- –≠—Ç–æ –ª–æ–≥–∏—á–Ω–æ, —Ç–∞–∫ –∫–∞–∫ –±–∞–∑–∞ —É –Ω–∞—Å –¥–æ –∏—é–Ω—è –º–µ—Å—è—Ü–∞.

3. –°–æ—Å—Ç–æ—è–Ω–∏–µ –¥–æ—Ä–æ–≥–∏ (road_surface)

- –¢–æ–ø ‚Äî –¥–Ω–µ–≤–Ω–æ–µ –≤—Ä–µ–º—è (~60%).
- –ù–æ—á—å —Å —Ñ–æ–Ω–∞—Ä—è–º–∏ ‚Äî –ø—Ä–∏–º–µ—Ä–Ω–æ —Ç—Ä–µ—Ç—å.

5. –°–≤–µ—Ç–æ—Ñ–æ—Ä—ã (control_device)
- –ë–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –∞–≤–∞—Ä–∏–π ‚Äî —Ç–∞–º, –≥–¥–µ –Ω–µ—Ç —Å–≤–µ—Ç–æ—Ñ–æ—Ä–∞.
- –†–∞–±–æ—á–∏–µ —Å–≤–µ—Ç–æ—Ñ–æ—Ä—ã –¥–∞—é—Ç –∑–∞–º–µ—Ç–Ω—É—é –≤—Ç–æ—Ä—É—é –≥—Ä—É–ø–ø—É.
- –ù–µ–∏—Å–ø—Ä–∞–≤–Ω—ã—Ö ‚Äî –º–∏–Ω–∏–º—É–º.

6. –ì–µ–æ–ª–æ–∫–∞—Ü–∏—è (county_city_location)
- –†–µ–≥–∏–æ–Ω Other –¥–∞—ë—Ç –±–æ–ª—å—à—É—é —á–∞—Å—Ç—å –¥–∞–Ω–Ω—ã—Ö.
- –û—Å—Ç–∞–ª—å–Ω—ã–µ ‚Äî —Ä–µ–¥–∫–∏–µ.

- –¥–ª—è –º–µ—Å—è—Ü–∞ –∏ —á–∞—Å–∞ —Å–æ–∑–¥–∞–¥–∏–º –∫–∞—Å—Ç–æ–º–Ω—ã–π —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥–µ—Ç –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å –≤—Ä–µ–º–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤ —Å–∏–Ω—É—Å-–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è;
- –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö –∑–∞–ø–æ–ª–Ω—è–µ–º 'unknown';
- –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö - -1;
- –∫–æ–¥–∏—Ä—É–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —á–µ—Ä–µ–∑ OneHotEncoder;
- –º–∞—Å—à—Ç–∞–±–∏—Ä—É–µ—Ç —á–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —á–µ—Ä–µ–∑ StandardScaler;

```python
class TimeFeatures(BaseEstimator, TransformerMixin):
    """
    –°–æ–∑–¥–∞—ë—Ç sin-–ø—Ä–∏–∑–Ω–∞–∫–∏ –∏–∑ –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö –ø–æ–ª–µ–π:
      - hour_sin  ‚Äî —Å—É—Ç–æ—á–Ω—ã–π —Ü–∏–∫–ª
      - day_sin  ‚Äî –¥–Ω–µ–≤–Ω–æ–π —Ü–∏–∫–ª
      - weekday_sin ‚Äî –Ω–µ–¥–µ–ª—å–Ω—ã–π —Ü–∏–∫–ª
    """
    def __init__(self, hour_col='hour', day_col='day', weekday_col='week_day'):
        self.hour_col = hour_col
        self.day_col = day_col
        self.weekday_col = weekday_col

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        df = X.copy()

        df['hour_sin'] = np.sin(2 * np.pi * df[self.hour_col] / 24)
        df['day_sin'] = np.sin(2 * np.pi * df[self.day_col] / 31)
        df['weekday_sin'] = np.sin(2 * np.pi * df[self.weekday_col] / 7)

        return df[['hour_sin', 'day_sin', 'weekday_sin']]
    
    def get_feature_names_out(self, input_features=None):
        return np.array(['hour_sin', 'day_sin', 'weekday_sin'])
```

```python
categorical = [
    'weather', 'road_surface', 'lighting',
    'control_device', 'direction', 'location_type',
    'vehicle_type', 'vehicle_transmission', 'county_city_location'
]

numeric = [
    'vehicle_age', 'insurance_premium', 'intersection'
]

time = ['hour', 'day', 'week_day']
```

```python
preprocessor = ColumnTransformer(
    transformers=[

        # –≤—Ä–µ–º–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —á–µ—Ä–µ–∑ –Ω–∞—à –∫–∞—Å—Ç–æ–º–Ω—ã–π —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä
        ('time',  Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='constant', fill_value=-1)),
            ('time', TimeFeatures()),
        ]), time),

        # –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ OneHotEncoder
        ('cat', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),
            ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False))
        ]), categorical),

        # —á–∏—Å–ª–æ–≤—ã–µ StandardScaler
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='constant', fill_value=-1)),
            ('scaler', StandardScaler())
        ]), numeric),
    ]
)
```

```python
X = crash.drop(columns=['at_fault'])
y = crash['at_fault']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)
```

–î–ª—è —É—á—ë—Ç–∞ –Ω–µ–ª–∏–Ω–µ–π–Ω—ã—Ö –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–µ–π —è –¥–æ–±–∞–≤–∏–ª–∞ —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å ‚Äî –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–∏–π –∞–Ω—Å–∞–º–±–ª–µ–≤—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º. –ö–∞–∫ –æ—Å–Ω–æ–≤–∞ - —Å–æ–≤—Ä–µ–º–µ–Ω–Ω–∞—è –±—É—Å—Ç–∏–Ω–≥–æ–≤–∞—è –º–æ–¥–µ–ª—å CatBoost, –∫–æ—Ç–æ—Ä–∞—è –æ–±—ã–Ω–æ —Ö–æ—Ä–æ—à–æ —Ä–∞–±–æ—Ç–∞—é—Ç —Å –¥–∞–Ω–Ω—ã–º–∏ —Å –Ω–µ–ª–∏–Ω–µ–π–Ω—ã–º–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—è–º–∏. –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –æ–±—ã–Ω–∞—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è. –¢–∞–∫–æ–π –Ω–∞–±–æ—Ä –æ–±–µ—Å–ø–µ—á–∏–≤–∞–µ—Ç —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –±–∞–∑–æ–≤—ã—Ö –∏ –ø—Ä–æ–¥–≤–∏–Ω—É—Ç—ã—Ö —Ä–µ—à–µ–Ω–∏–π, —á—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ–±—ä–µ–∫—Ç–∏–≤–Ω–æ –≤—ã–±—Ä–∞—Ç—å –æ–ø—Ç–∏–º–∞–ª—å–Ω—É—é –º–æ–¥–µ–ª—å.

–î–ª—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ —è –±—É–¥—É –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å SelectKbest.        
–í –∫–∞—á–µ—Å—Ç–≤–µ –æ—Å–Ω–æ–≤–Ω–æ–π –º–µ—Ç—Ä–∏–∫–∏ - F1, –ø–æ—Ç–æ–º—É —á—Ç–æ –≤ –∫–∞—Ä—à–µ—Ä–∏–Ω–≥–µ –≤–∞–∂–Ω—ã –∏ –ª–æ–∂–Ω—ã–µ –ø—Ä–æ–ø—É—Å–∫–∏ –∞–≤–∞—Ä–∏–π, –∏ –ª–æ–∂–Ω—ã–µ –æ–±–≤–∏–Ω–µ–Ω–∏—è –≤–æ–¥–∏—Ç–µ–ª–µ–π:
- –õ–æ–∂–Ω–æ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ - –º–æ–¥–µ–ª—å –ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç –≤–∏–Ω–æ–≤–Ω—ã—Ö - –∫–æ–º–ø–∞–Ω–∏—è —Ç–µ—Ä—è–µ—Ç –¥–µ–Ω—å–≥–∏, —Å—Ç—Ä–∞—Ö–æ–≤—ã–µ –≤—ã–ø–ª–∞—Ç—ã —Ä–∞—Å—Ç—É—Ç.
- –õ–æ–∂–Ω–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ -  –º–æ–¥–µ–ª—å –æ—à–∏–±–æ—á–Ω–æ –ø–æ–º–µ—á–∞–µ—Ç –Ω–µ–≤–∏–Ω–æ–≤–Ω—ã—Ö - –∂–∞–ª–æ–±—ã, —É—Ö—É–¥—à–µ–Ω–∏–µ –æ–ø—ã—Ç–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤, —Ä–µ–ø—É—Ç–∞—Ü–∏–æ–Ω–Ω—ã–µ —Ä–∏—Å–∫–∏.
- F1 —É–¥–µ—Ä–∂–∏–≤–∞–µ—Ç –±–∞–ª–∞–Ω—Å –º–µ–∂–¥—É —Ç–µ–º, —á—Ç–æ–±—ã –Ω–µ –ø—Ä–æ–ø—É—Å–∫–∞—Ç—å –≤–∏–Ω–æ–≤–Ω—ã—Ö, –∏ —Ç–µ–º, —á—Ç–æ–±—ã –Ω–µ –æ–±–≤–∏–Ω—è—Ç—å –ª–∏—à–Ω–∏—Ö.

```python
 #--- –ú–æ–¥–µ–ª–∏ ---
models = {
    'RandomForest': (
        RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1),
        preprocessor
    ),
    'LogisticRegression': (
        LogisticRegression(random_state=RANDOM_STATE, max_iter=2000, solver='liblinear'),
        preprocessor
    ),
    'CatBoost': (
        CatBoostClassifier(random_state=RANDOM_STATE, verbose=0),
        preprocessor
    )
}

# --- –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è GridSearchCV ---
param_distributions = {
    'RandomForest': {
        'select__k': [10, 15, 20, 25],
        'model__n_estimators': [100, 150, 200],
        'model__max_depth': [5, 10, 15],
        'model__min_samples_split': [2, 5, 10]
    },
    'CatBoost': {
        'select__k': [10, 15, 20, 25],
        'model__depth': [5, 7, 9],
        'model__learning_rate': [0.01, 0.05, 0.1],
        'model__iterations': [100, 150, 200]
    },
    'LogisticRegression': {
        'select__k': [10, 15, 20, 25],
        'model__C': uniform(0.01, 10)
    }
}

# --- –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π ---
results = []
results_models = {}

for name, (model, preprocessor) in models.items():
    print(f'\nüîπ –û–±—É—á–∞–µ–º {name}...')

    pipe = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('select', SelectKBest(score_func=f_classif)),
        ('model', model)
    ])
    
    scoring = {
        'f1': 'f1',
        'roc_auc': 'roc_auc'
    }

    search = RandomizedSearchCV(
        pipe,
        param_distributions[name],
        scoring=scoring,
        refit='f1',
        cv=3,
        n_iter=5,
        n_jobs=-1,
        random_state=RANDOM_STATE,
        verbose=2
    )

    search.fit(X_train, y_train)

    best_model = search.best_estimator_
    results_models[name] = best_model
    
    cv_roc_auc = search.cv_results_['mean_test_roc_auc'][search.best_index_]

    
    results.append({
        'Model': name,
        'CV_f1': search.best_score_,
        'CV_roc_auc': cv_roc_auc,
        'Best_Params': search.best_params_
    })

results_df = pd.DataFrame(results).sort_values('CV_f1', ascending=False)
print('\033[1m\n–ò—Ç–æ–≥–æ–≤–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π:\033[0m')
results_df
```
